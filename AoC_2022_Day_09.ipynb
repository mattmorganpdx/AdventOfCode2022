```python
# Set up
move_map ={
    'R': lambda x: (x[0], x[1] + 1),
    'L': lambda x: (x[0], x[1] - 1),
    'U': lambda x: (x[0] + 1, x[1]),
    'D': lambda x: (x[0] - 1, x[1]),
}

def not_adj(H, T):
  for x, y in zip(H, T):
    if abs(x - y) > 1:
      return True
  return False

def new_loc(K, N):
  return *(y if x == y else y + ((x - y)//abs(x - y)) for x, y in zip(K, N)),
```

```python
# Part 1
H = (0,0)
T = (0,0)
T_path = set((0,0))

for line in input.splitlines():
  k, m = line.split()
  for _ in range(int(m)):
    prev = H
    H = move_map.get(k)(H)
    if not_adj(H, T):
      T = prev
      T_path.add(T)

len(T_path)
```

```python
# Part 1
knots = [(0,0)] * 10
T_path = set()

for line in input.splitlines():
  k, m = line.split()
  for _ in range(int(m)):
    knots[0] = move_map.get(k)(knots[0])
    for knot in range(9):
      if not_adj(knots[knot], knots[knot + 1]):
        knots[knot + 1] = new_loc(knots[knot], knots[knot + 1])
    T_path.add(knots[-1])

len(T_path)
```
