```python
# Setup
import re

rock_match = r'(\d+,\d+)'

def get_rock_locs(i):
  rock_locs = set()

  for line in i.splitlines():
    rng = list()
    for m in re.findall(rock_match, line):
      x, y = map(int, m.split(','))
      rng.append((x,y))
    for idx, i in enumerate(rng[:-1], 1):
      if i[0] == rng[idx][0]:
        if i[1] < rng[idx][1]:
          start, stop = i[1], rng[idx][1]
        else:
          start, stop = rng[idx][1], i[1]
        rock_locs.update({(i[0], y) for y in range(start, stop +1)})
      elif i[1] == rng[idx][1]:
        if i[0] < rng[idx][0]:
          start, stop = i[0], rng[idx][0]
        else:
          start, stop = rng[idx][0], i[0]
        rock_locs.update({(x, i[1]) for x in range(start, stop +1)})
      else:
        print(f"ERROR somethings not right with {idx, i}")
    
  return rock_locs
```
