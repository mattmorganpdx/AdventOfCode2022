```python
import re

def build_tree(input):
  # This regex parse matches cd and ls commands.
  # items that come out of re.findall will match one or the other.
  input_parse_re = r'((?:(?<=ls\n)[^$]*)|(?:cd (?:\.\.|\w+|/)))'

  # The tree is a flat map of fully qualified path to contents.
  tree = {'/': []}

  # We need to treat cwd as a stack because we need the key
  # in the map to be unique so we'll join cwd with '/' like real life.
  cwd = []
  for line in re.findall(input_parse_re, input):
    line = line.strip()
    # Handle cd
    if line.startswith('cd '):
      path = line.split()[1]
      if path == '/':
        cwd = ['/']
      elif path == '..':
        cwd = cwd[:-1]
      else:
        cwd.append(path)
    # Handle ls, the regex pretrimed the 'ls\n' off for us.
    else:
      tree['/'.join(cwd)] = [file for file in line.splitlines()]
  return tree
    
# dict to mutate for tracking the sizes
dir_sizes = {}

def get_dir_size(dir, tree):
  files = tree[dir]
  size = sum(get_dir_size(f'{dir}/{file.split()[1]}', tree) if file.startswith('dir') else int(file.split()[0]) for file in files)
  dir_sizes[dir] = size
  return size

tree = build_tree(input)

get_dir_size('/', tree)
```

```python
# Part 1
sum(size for size in dir_sizes.values() if size <= 100000)
```

```python
# Part 2
free_space_needed = 30000000 - (70000000 - dir_sizes['/'])
min(size for size in dir_sizes.values() if size >= free_space_needed)
```
