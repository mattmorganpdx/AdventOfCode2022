```python
import re
from queue import LifoQueue

def build_queues(raw_stacks):
  # Build Queues based on last column number
  queues = []
  for _ in range(int(raw_stacks.splitlines()[-1].rstrip()[-1])):
    queues.append(LifoQueue())

  # Regex to break up the crate line
  crate_line_match = r'(?:(.{3})\s?)'

  # Populate Queues
  for row in raw_stacks.splitlines()[-2::-1]:
    for idx, crate in enumerate(re.findall(crate_line_match, row)):
      if crate != '   ':
        queues[idx].put(crate)
  
  return queues

# Regex to break up the move line
move_line_match = r'move (\d+) from (\d+) to (\d+)'
```

```python
# Part 1
raw_stacks, moves = input.split('\n\n')

queues = build_queues(raw_stacks)

for move in moves.splitlines():
  count, _from, to = map(int, re.match(move_line_match, move).groups())
  for _ in range(count):
    queues[to - 1].put(queues[_from - 1].get())

''.join(queue.get() for queue in queues).replace('[', '').replace(']', '')
```

```python
# Part 2
raw_stacks, moves = input.split('\n\n')

queues = build_queues(raw_stacks)

for move in moves.splitlines():
  count, _from, to = map(int, re.match(move_line_match, move).groups())
  temp = []
  for _ in range(count):
    crate = queues[_from - 1].get()
    temp.append(crate)
  for crate in temp[::-1]:
    queues[to - 1].put(crate)

''.join(queue.get() for queue in queues).replace('[', '').replace(']', '')
```
