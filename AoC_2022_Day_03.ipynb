```python
# Part 1
import string

def get_common_item(pack):
  half = len(pack) // 2
  common_item, = set(pack[:half]).intersection(pack[half:])
  return string.ascii_letters.index(common_item) + 1

sum(get_common_item(pack) for pack in input.splitlines())
```
