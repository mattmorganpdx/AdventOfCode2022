```python
trans = {'-': '-1', '=': '-2'}

def get_total_fuel_reqs(pi):
  total = 0
  for line in pi.splitlines():
    dec = 0
    for place, num in enumerate(line[::-1]):
      num = int(trans.get(num, num))
      dec += (5 ** place) * num
    total += dec
  return total

def dec_to_snafu(n):
  char_map = {
      0: '0',
      1: '1',
      2: '2',
      3: '=',
      4: '-',
      5: '0'
  }
  m = ""
  c = 0
  # c is to carry forward
  # if r is > 2 then sub the correct repr and set the carry
  while n != 0:
    n, r = divmod(n, 5)
    r, c = r + c, 0
    if r > 2:
      c = 1

    m = char_map.get(r) + m

  return m


dec_to_snafu(get_total_fuel_reqs(pi))
```
