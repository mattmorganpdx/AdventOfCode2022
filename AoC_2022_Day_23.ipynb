```python
from collections import deque

def draw_locations(elf_locations):
  min_y = min(elf_locations)[0]
  max_y = max(elf_locations)[0]
  min_x = min(l[1] for l in elf_locations)
  max_x = max(l[1] for l in elf_locations)
  count = 0
  for y in range(min_y, max_y + 1):
    for x in range(min_x, max_x + 1):
      print('#' if (y,x) in elf_locations else '.', end="")
      count += 0 if (y,x) in elf_locations else 1
    print()
  print(f"found {count} empty dots")

elf_locations = []

for idx, line in enumerate(pi.splitlines()):
  for jdx, space in enumerate(line):
    if space == '#':
      elf_locations.append((idx, jdx))

def check_empty(elf, f):
  for i in [-1, 0, 1]:
    if f(elf, i) in elf_locations:
      return False
  return True

direction_check_functions = {
    'n': lambda elf, i: (elf[0] - 1, elf[1] + i),
    's': lambda elf, i: (elf[0] + 1, elf[1] + i),
    'e': lambda elf, i: (elf[0] + i, elf[1] + 1),
    'w': lambda elf, i: (elf[0] + i, elf[1] - 1),
}
move_functions = {
    'n': lambda elf: (elf[0] - 1, elf[1]),
    's': lambda elf: (elf[0] + 1, elf[1]),
    'e': lambda elf: (elf[0], elf[1] + 1),
    'w': lambda elf: (elf[0], elf[1] - 1)
}

directions = deque(['n', 's', 'w', 'e'])

def round_part_1(directions):
  suggested_moves = {}
  for i, elf in enumerate(elf_locations):
    if not all(check_empty(elf, direction_check_functions.get(d)) for d in directions):
      for d in directions:
        if check_empty(elf, direction_check_functions.get(d)):
          if not suggested_moves.get(elf):
            suggested_moves[elf] = move_functions.get(d)(elf)
  return suggested_moves

def round_part_2(suggested_moves):
  move_count = 0
  for i, elf in enumerate(elf_locations):
    suggestion = suggested_moves.get(elf)
    if list(suggested_moves.values()).count(suggestion) == 1:
      elf_locations[i] = suggestion
      move_count += 1
  return move_count
    

for _ in range(10):
  round_part_2(round_part_1(directions))
  directions.rotate(-1)

# Part 1
draw_locations(elf_locations)
round = 11

while round_part_2(round_part_1(directions)):
  directions.rotate(-1)
  round += 1
  
print(round)
```
