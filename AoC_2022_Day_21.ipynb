```python
# Part 1
monkeys = {}

pi = test_input.splitlines()

while not monkeys.get('root'):
  for idx, m in enumerate(pi):
    name, action = m.split(": ")
    try:
      value = int(action)
      monkeys[name] = str(value)
    except:
      new_line = ' '.join(list(map(lambda x: monkeys.get(x,x), action.split())))
      try:
        value = eval(new_line)
        monkeys[name] = str(value)
      except:
        pi[idx] = f"{name}: {new_line}"

monkeys.get('root')
```

```python
# Part 2
from sympy import Symbol, solve

pi = input.splitlines()

# Create a map between var and formula
res_dict = {ww.split(': ')[0]: ww.split(': ')[1] for ww in pi}

# Recursive search to resolve everything but 'humn' which we'll replace with x
def find(look_for):
  if look_for == 'humn':
    return 'x'
  v = res_dict.get(look_for)
  try:
    return str(int(v))
  except:
    a, o, b = v.split()
    e = f"({find(a)} {o} {find(b)})"
    try:
      return str(eval(e))
    except:
      return e

# Replace the op in the root formula to be a '=='
res_dict['root'] = res_dict.get('root').replace('+', '==')

# Get each have of the equality
l , r = find('root').split('==')

# Solve for 'humn'
solve(f"{l} - {r}", Symbol('x'))
```
